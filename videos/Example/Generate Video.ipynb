# Generate Video 

Takes the folders full of images and generates a video animating the loss landscapes over time.

```python
import time

from minima_volume.landscape_videos import (
    create_video_ffmpeg
)
```

```python
# Create videos using matplotlib built-ins only and time each one
print("\n=== Creating Videos ===")

plot_dirs = {
    'dynamic_vmin': 'plots_dynamic_vmin',
    'fixed_vmin_large': 'plots_fixed_vmin_large', 
    'fixed_vmin_small': 'plots_fixed_vmin_small', 
    '3d': 'plots_3d',
    'videos': 'videos'
}

video_configs = [
    ('dynamic_vmin', 'loss_landscape_dynamic_vmin.mp4'),
    ('fixed_vmin_small', 'loss_landscape_fixed_vmin_small.mp4'),
    ('fixed_vmin_large', 'loss_landscape_fixed_vmin_large.mp4'),
    ('3d', 'loss_landscape_3d.mp4')
]

for folder_key, filename in video_configs:
    start_time = time.time()
    print(f"Creating {filename}...")
    
    create_video_ffmpeg(
        image_folder=plot_dirs[folder_key],
        output_video_path=f"{plot_dirs['videos']}/{filename}",
        fps=5
    )
    
    end_time = time.time()
    elapsed_time = end_time - start_time
    print(f"✓ {filename} created in {elapsed_time:.2f} seconds")

print("\n✓ All videos created successfully!")
```
