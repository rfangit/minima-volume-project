```python
import matplotlib.pyplot as plt
import numpy as np

# Set up the plot
plt.figure(figsize=(6, 6))

# Define your lines here - each line is defined by slope, intercept, and points
# Format: {'slope': m, 'intercept': b, 'points': [(x1, y1), (x2, y2), ...], 'label': 'Line Name', 'color': 'color_name'}
lines = [
    {
        'slope': -10,
        'intercept': 5,
        'points': [(0, 5)],#, (2, 5), (3, 7)],
        'label': 'Line 1',
        'color': 'blue'
    },
    {
        'slope': -1.5,
        'intercept': 4.5,
        'points': [(1, 3)],#, (2, 4), (3, 5)],
        'label': 'Line 2', 
        'color': 'red'
    },
    {
        'slope': -0.5,
        'intercept': 3,
        'points': [(2, 2)],#, (2, 4), (3, 5)],
        'label': 'Line 3', 
        'color': 'purple'
    },
        {
        'slope': -0.2,
        'intercept': 2.2,
        'points': [(4, 1.4)],#, (2, 4), (3, 5)],
        'label': 'Line 3', 
        'color': 'green'
    },
    # Add more lines as needed
]

# X-axis range for plotting lines (log scale)
x_min, x_max = 0, 5
x_values = np.linspace(x_min, x_max, 100)

# Plot each line
for line in lines:
    # Calculate y values for the line
    y_values = line['slope'] * x_values + line['intercept']
    
    # Plot the line
    plt.plot(x_values, y_values, 
             #label=line['label'], 
             color=line['color'],
             linestyle='-',
             linewidth=2)
    
    # Plot the points
    points = line['points']
    x_points = [p[0] for p in points]
    y_points = [p[1] for p in points]
    plt.scatter(x_points, y_points, 
                color=line['color'], 
                s=300,  # marker size
                zorder=5)  # bring points to front

# Customize the plot
plt.xlabel('Log (Dataset Size)', fontsize=30)
plt.ylabel('Volume', fontsize=30)
plt.title('Volume vs Log(Dataset Size)', fontsize=30)
plt.grid(True, alpha=0.3)
#plt.legend()

# Set x-axis to show log scale (you can adjust the ticks as needed)
plt.xticks(ticks=[0, 1, 2, 3, 4, 5], 
           labels=['$10^0$', '$10^1$', '$10^2$', '$10^3$', '$10^4$', '$10^5$'])
plt.ylim([0, 6])

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data - red and green points only
red_x = [-1, 1, -0.2, 1.2]
red_y = [1, 1, -1, -0.8]
red_size = [7000, 7000, 7000, 7000]

green_x = [0]
green_y = [0]
green_size = [10000]

# Create a single figure and axis
fig, ax = plt.subplots(figsize=(6, 6))

# Plot green circles
ax.scatter(green_x, green_y, s=green_size, c='green', edgecolors='black',
           linewidth=1, alpha=0.7)

# Plot red circles
ax.scatter(red_x, red_y, s=red_size, c='red', edgecolors='black',
           linewidth=1, alpha=0.7)

# Set title and axis formatting
ax.set_title("Weak Volume Hypothesis", fontsize=20)
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.grid(True, alpha=0.3)

# Fix axis limits
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)

# Main figure title
fig.suptitle("Volume Hypothesis - Loss Landscapes", fontsize=25, fontweight='bold', y=1.05)

# Optional: Legend (just for red/green)
from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w', markerfacecolor='green',
           markersize=10, markeredgecolor='black', linewidth=1, label='Low Test Loss'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='red',
           markersize=10, markeredgecolor='black', linewidth=1, label='High Test Loss'),
]
fig.legend(handles=legend_elements, loc='lower center',
           bbox_to_anchor=(0.5, 0.91), ncol=2, frameon=True,
           fancybox=True, shadow=True, fontsize=12)

plt.tight_layout()
#plt.savefig("red_green_points.png", bbox_inches='tight', dpi=300,
#            facecolor='white', edgecolor='none')
plt.show()
```
