```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import numpy as np

# Sample data - replace these with your actual data
# Plot 1 data
red_x_loc_1 = [-1, 1, 0]#, 3, 5]
red_y_loc_1 = [1, 1, -1]#, 4, 6]
red_size_1 = [25000, 2000, 900]#, 200, 150]

green_x_loc_1 = [0]#, 2, -2]
green_y_loc_1 = [0]#, 2, -2]
green_size_1 = [1200]#, 180, 90]

purple_x_loc_1 = [-1, -1.5, 0.8, 1.0, 0.3]#-1, 3, -3]
purple_y_loc_1 = [-1, -0.5, -1.5, -0.2, 1.0]#2, 1.5, 0]
purple_size_1 = [150, 150, 150, 150, 150]#80, 160, 140]

# Plot 2 data
red_x_loc_2 = [-1.5, -0.8, 1, 0]#, 3, 5]
red_y_loc_2 = [0.5, 1.2, 1, -1]#, 4, 6]
red_size_2 = [500, 500, 500, 500]#, 200, 150]

green_x_loc_2 = [0]#, 2, -2]
green_y_loc_2 = [0]#, 2, -2]
green_size_2 = [1000]#, 180, 90]

purple_x_loc_2 = [-1, -1.5, 0.8, 1.0, 0.3]#-1, 3, -3]
purple_y_loc_2 = [-1, -0.5, -1.5, -0.2, 1.0]#2, 1.5, 0]
purple_size_2 = [150, 150, 150, 150, 150]#80, 160, 140]

# Plot 3 data
#red_x_loc_2 = [-1.5, -0.8, 1, 0]#, 3, 5]
#red_y_loc_2 = [0.5, 1.2, 1, -1]#, 4, 6]
#red_size_2 = [10, 10, 10, 10]#, 200, 150]

green_x_loc_3 = [0]#, 2, -2]
green_y_loc_3 = [0]#, 2, -2]
green_size_3 = [800]#, 180, 90]

purple_x_loc_3 = [-1, -1.5, 0.8, 1.0, 0.3]#-1, 3, -3]
purple_y_loc_3 = [-1, -0.5, -1.5, -0.2, 1.0]#2, 1.5, 0]
purple_size_3 = [150, 150, 150, 150, 150]#80, 160, 140]

# Create figure with 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 6))

def plot_circles(ax, red_x, red_y, red_size, green_x, green_y, green_size, purple_x, purple_y, purple_size, title):
    # Plot green circles
    ax.scatter(green_x, green_y, s=green_size, c='green', edgecolors='black', linewidth=1, alpha=0.7)
    
    # Plot red circles
    ax.scatter(red_x, red_y, s=red_size, c='red', edgecolors='black', linewidth=1, alpha=0.7)
    
    # Plot purple circles
    ax.scatter(purple_x, purple_y, s=purple_size, c='purple', edgecolors='black', linewidth=1, alpha=0.7)
    
    ax.set_title(title, fontsize = 20)
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.grid(True, alpha=0.3)
    
    # Set fixed axis limits from -2 to 2
    ax.set_xlim(-2, 2)
    ax.set_ylim(-2, 2)
    
# Create custom legend with fixed size markers
legend_elements = [
    Line2D([0], [0], marker='o', color='w', markerfacecolor='green', 
           markersize=10, markeredgecolor='black', linewidth=1, label='Low Test Loss (Generalization)'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='red', 
           markersize=10, markeredgecolor='black', linewidth=1, label='High Test Loss (Memorization)'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='purple', 
           markersize=10, markeredgecolor='black', linewidth=1, label='High Test Loss (Poisoned Data)')
]

# Your plotting code remains the same...
plot_circles(axes[2], red_x_loc_1, red_y_loc_1, red_size_1, 
             green_x_loc_1, green_y_loc_1, green_size_1, 
             purple_x_loc_1, purple_y_loc_1, purple_size_1, 
             title="High Data Regime")

plot_circles(axes[1], red_x_loc_2, red_y_loc_2, red_size_2, 
             green_x_loc_2, green_y_loc_2, green_size_2, 
             purple_x_loc_2, purple_y_loc_2, purple_size_2, 
             title="Grokking Regime")

plot_circles(axes[0], red_x_loc_3, red_y_loc_3, red_size_3, 
             green_x_loc_3, green_y_loc_3, green_size_3, 
             purple_x_loc_3, purple_y_loc_3, purple_size_3, 
             title="Low Data Regime")

# Add main title above all figures
fig.suptitle("Volume Hypothesis - Loss Landscapes Across Different Data Regimes", fontsize=25, fontweight='bold', y=1.05)

# Create a separate legend above all plots
fig.legend(handles=legend_elements, loc='lower center', 
           bbox_to_anchor=(0.5, 0.91), ncol=3, frameon=True, 
           fancybox=True, shadow=True, fontsize=12)

plt.subplots_adjust(top=0.80)  # Give more space if needed
plt.tight_layout()
# Save the figure with high quality and proper bounding box
plt.savefig("volume_hypothesis_cartoon.png", bbox_inches='tight', dpi=300, facecolor='white', edgecolor='none')
plt.show()
```
