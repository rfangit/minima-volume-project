```python
import os
import shutil

# names you want to delete
TARGETS = {"loss", "accuracy", "log_volume.png", "log_volume_generalization.png", "model_modification_vs_test_loss.png", "results.json"}  

# get current folder (where this script is located)
base_dir = "."

for root, dirs, files in os.walk(base_dir, topdown=False):
    # check files
    for name in files:
        if name in TARGETS:
            path = os.path.join(root, name)
            os.remove(path)
            print(f"Deleted file: {path}")

    # check directories
    for name in dirs:
        if name in TARGETS:
            path = os.path.join(root, name)
            shutil.rmtree(path, ignore_errors=True)
            print(f"Deleted folder: {path}")
```
