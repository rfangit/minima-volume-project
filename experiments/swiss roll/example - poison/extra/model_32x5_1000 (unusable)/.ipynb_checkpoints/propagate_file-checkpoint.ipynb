```python
import os
import shutil
import sys

# === USER CONFIGURATION ===
source_folder = "base"
notebook_name = "Random Perturbs.ipynb"

# Path to the master notebook
source_file = os.path.join(source_folder, notebook_name)

# Check if source notebook exists
if not os.path.exists(source_file):
    print(f'ERROR: Master notebook not found at "{source_file}"')

# Get all model_* subfolders
exclude_starts = ['.ipynb_checkpoints', 'Example', 'Finished', 'Not Started']

target_folders = [d for d in os.listdir('.') 
           if os.path.isdir(d) and not any(d.startswith(prefix) for prefix in exclude_starts)]

if not target_folders:
    print("No model_* folders found in the current directory.")

# Copy notebook to each target folder
copied_count = 0
for folder in target_folders:
    try:
        target_path = os.path.join(folder, notebook_name)
        shutil.copy2(source_file, target_path)
        print(f'Copied {notebook_name} into {folder}...')
        copied_count += 1
    except Exception as e:
        print(f'ERROR: Failed to copy to {folder}: {e}')

print(f'\nDone propagating {notebook_name} to {copied_count} model_* folders.')
```
