```python
import os
import subprocess
import time

# Notebooks to run in order
NOTEBOOKS = [
    "make_dataset_notebooks.ipynb",
    "run_random_perturbs_parallel.ipynb",
    "run_volume_parallel.ipynb", 
]

def run_notebook(nb):
    if not os.path.exists(nb):
        print(f"‚ùå {nb} not found, skipping.")
        return
    print(f"üîπ Running {nb}...")
    start = time.time()
    subprocess.run([
        "jupyter", "nbconvert", "--to", "notebook",
        "--execute", nb, "--inplace",
        "--ExecutePreprocessor.timeout=-1"
    ], check=True)
    elapsed = time.time() - start
    print(f"‚úÖ Finished {nb} (took {elapsed:.2f}s)\n")

for nb in NOTEBOOKS:
    run_notebook(nb)

print("üéâ All notebooks done!")
```
