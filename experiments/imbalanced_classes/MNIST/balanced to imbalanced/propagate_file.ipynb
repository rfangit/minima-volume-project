```python
import os
import shutil

# === USER CONFIGURATION ===
source_folder = "base folder"
files_to_copy = [
    "Random Perturbs.ipynb",
    "Volume Estimation Pipeline.ipynb",
    "Volume Cutoff.ipynb",
    # some model code 
    "swiss_model_data.py"
]

# Check if files exist in source
missing_files = [f for f in files_to_copy if not os.path.exists(os.path.join(source_folder, f))]
if missing_files:
    print("WARNING: These files were not found in base:", missing_files)

# Get all target folders
EXCLUDE_PREFIXES = ['.ipynb_checkpoints', 'base', 'completed runs', 'to do', 'analysis']

target_folders = [
    d for d in os.listdir('.')
    if os.path.isdir(d) and not any(d.startswith(prefix) for prefix in EXCLUDE_PREFIXES)
]

if not target_folders:
    print("No model_* folders found in the current directory.")

# Copy each file to each target folder
copied_total = 0
for folder in target_folders:
    for file_name in files_to_copy:
        source_path = os.path.join(source_folder, file_name)
        target_path = os.path.join(folder, file_name)

        if not os.path.exists(source_path):
            continue  # skip missing files silently (already warned above)

        try:
            shutil.copy2(source_path, target_path)
            print(f'Copied {file_name} → {folder}')
            copied_total += 1
        except Exception as e:
            print(f'ERROR: Failed to copy {file_name} to {folder}: {e}')

print(f'\nDone propagating {len(files_to_copy)} files into {len(target_folders)} folders.')
print(f'Total copy operations: {copied_total}')
```
