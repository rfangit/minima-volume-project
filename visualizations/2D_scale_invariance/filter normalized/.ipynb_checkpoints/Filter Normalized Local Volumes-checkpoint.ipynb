# Introduction

Previously, we calculated the volume (for $a > w$ as

$$
\begin{split}
V &= 2\sqrt{2w (a + w)} + 2w \log \frac{\left(1 + \sqrt{\frac{w}{a+w}} \right)}{\left(1 - \sqrt{\frac{w}{a+w}} \right)} - (a - w)\log \frac{\left(1 + \sqrt{\frac{2w}{a + w}} \right)}{\left(1 - \sqrt{\frac{2w}{a + w}} \right)}
\end{split}
$$

Now we consider the effects of filter normalizing. 

## Filter Normalization

Recall that our point was originally $(\sqrt{a}b, \sqrt{a}/b)$. We shoot out random vectors $\vec{r}$ and track when they reach the boundaries. The effect of filter normalization is that the vector r is transformed from the original $\vec{r} = v_x, v_y$ to $\vec{r}_{norm} = v_x \times \sqrt{a}b, v_y \times \sqrt{a}/b$.

This warps the original set we computed the volume of. Specifically, the height of the shape becomes $h_{norm} = h_{old} \times b/\sqrt{a}$, $w_{norm} = w_{old} / (\sqrt{a}b)$. The total volume should be given by $V_{norm} = V_{old}/ a$.

## Overview

Applying this algorithm, we find filter normalization doesn't fix any of the issues with star convexity (nor the fundamental issues with scale invariance), but it does even out the numerical effects of scale invariance.

In fact, it seems like the filter normalization actually makes our star convexity problem worse. But this may be an artifact of the system we study (where the total normal of the weights has such a huge impact). 

Overall, it seems reasonable to use filter normalization, but also try runs without it.
