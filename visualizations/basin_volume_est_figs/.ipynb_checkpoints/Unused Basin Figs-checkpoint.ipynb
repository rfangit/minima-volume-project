# SGD vs Guess And Check

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

def super_gaussian(x, y, x0, y0, sigma=1.0, amplitude=1.0, power=4):
    """Super-Gaussian with adjustable power (4=quartic dropoff)"""
    r_sq = (x-x0)**2 + (y-y0)**2
    return amplitude * np.exp(-(r_sq / (2*sigma**2)) ** (power/2))

# Create custom white-to-red colormap
colors = [(1, 1, 1), (1, 0, 0)]  # White to Red
cmap = LinearSegmentedColormap.from_list('white_red', colors)

# Create grid with higher resolution for smoother visualization
x = np.linspace(-7, 7, 40)
y = np.linspace(-7, 7, 40)
X, Y = np.meshgrid(x, y)

# Define potentials
centers = [(-3, -3), (3, 3)]
sigmas = [1.5, 1]
amplitudes = [1.0, 1.0]  # More negative = deeper wells
powers = [8, 2]  # Higher power = sharper dropoff

# Compute fields
Z = np.zeros_like(X)
for (x0, y0), sigma, amp, p in zip(centers, sigmas, amplitudes, powers):
    Z += super_gaussian(X, Y, x0, y0, sigma, amp, p)

## Now the vector field
xv = np.linspace(-7, 7, 16)
yv = np.linspace(-7, 7, 16)
Xv, Yv = np.meshgrid(xv, yv)

U, V = np.zeros_like(Xv), np.zeros_like(Yv)
for (x0, y0), sigma, amp, p in zip(centers, sigmas, amplitudes, powers):
    r_sq = (Xv-x0)**2 + (Yv-y0)**2
    g = super_gaussian(Xv, Yv, x0, y0, sigma, amp, p)
    U += -p * (Xv-x0) * (r_sq ** (p/2 - 1)) / (2*sigma**2)**(p/2) * g
    V += -p * (Yv-y0) * (r_sq ** (p/2 - 1)) / (2*sigma**2)**(p/2) * g

# Normalize each vector to unit length
magnitude = np.sqrt(U**2 + V**2)
# Avoid division by zero for zero vectors
#magnitude[magnitude == 0] = 1  # Set zero magnitudes to 1 (will result in zero vectors)
U_norm = U / (magnitude)
V_norm = V / (magnitude)

scaled_magnitude = (0.4 + magnitude)
U_scaled = U_norm * scaled_magnitude
V_scaled = V_norm * scaled_magnitude

# Create figure
plt.figure(figsize=(12, 9))

# Plot potential field with white-to-red colormap
# Set vmin to slightly below minimum value to ensure white background
plt.imshow(Z, extent=(-7, 7, -7, 7), origin='lower', 
           cmap=cmap, alpha=0.8)#, vmin=np.min(Z)-0.1, vmax=0)
cbar = plt.colorbar(label='Potential Value')
#cbar.set_ticks([np.min(Z), 0])  # Show min and max values

# Plot vector field with:
# - scale=15 (smaller = larger arrows)
# - width=0.005 (thicker arrows)
# - color='darkblue' for better contrast
plt.quiver(Xv, Yv, U_scaled, V_scaled, scale=20, width=0.003, 
           color='darkblue', pivot='middle')

#plt.title('Potential Field and Gradient Vectors\n(White=High Potential, Red=Low Potential)')
#plt.xlabel('X position')
#plt.ylabel('Y position')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

# Generic Presentation Graphics

```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian(x, mu, sigma, amplitude):
    """Negative Gaussian function."""
    return -amplitude * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

def sum_of_gaussians(x, centers, sigma, amplitude, center_sigma=None, center_amplitude=None):
    """Sum of Gaussians with special center Gaussian."""
    total = 0.0
    for i, center in enumerate(centers):
        if i == len(centers) // 2 and center_sigma is not None and center_amplitude is not None:
            total += gaussian(x, center, center_sigma, center_amplitude)  # Special center
        else:
            total += gaussian(x, center, sigma, amplitude)  # Regular Gaussians
    return total

# Parameters
num_gaussians = 3
amplitude = 1.0
variance = 0.2
sigma = np.sqrt(variance)
spacing = 6.0
x_min, x_max = -12, 12

# Special center Gaussian
center_amplitude = 1.
center_variance = 0.2
center_sigma = np.sqrt(center_variance)

# Centers (equally spaced)
centers = np.linspace(-spacing * (num_gaussians - 1) / 2, 
                      spacing * (num_gaussians - 1) / 2, 
                      num_gaussians)

# Generate x values
x = np.linspace(x_min, x_max, 1000)

# Full sum (black line)
y = sum_of_gaussians(x, centers, sigma, amplitude, center_sigma, center_amplitude)

# --- Define the center region to highlight ---
threshold_distance = 2.0  # Distance from center to switch to red
center_mu = centers[num_gaussians // 2]  # Center of middle Gaussian
mask = (x >= center_mu - threshold_distance) & (x <= center_mu + threshold_distance)

# Split into 3 segments: left (black), center (red), right (black)
x_left = x[~mask & (x < center_mu)]
y_left = y[~mask & (x < center_mu)]
x_right = x[~mask & (x > center_mu)]
y_right = y[~mask & (x > center_mu)]
x_center = x[mask]
y_center = y[mask]

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(x_left, y_left, 'k-', linewidth=3, label='Memorization Minima')  # Left black
plt.plot(x_right, y_right, 'k-', linewidth=3)  # Right black
plt.plot(x_center, y_center, 'r-', linewidth=3, label='Generalization Minima')  # Center red
plt.ylim(-1.8, 1)
#plt.xlabel('x', fontsize=14)
#plt.ylabel('Loss', fontsize=14)
plt.xticks([])
plt.yticks([])
plt.title("Loss Landscape", fontsize = 30)
plt.legend(fontsize=20)
plt.grid(True, alpha=0.2)
plt.savefig('figs/idea_1_low_data.png', dpi=300, bbox_inches='tight', transparent=False)
plt.show()
```

# Fundamental Volume Metrics

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.path import Path

def create_bumpy_spheroid(center, a, b, theta, bumpiness=0.1, num_bumps=8):
    """Generate points for a spheroid with smooth bumps"""
    t = np.linspace(0, 2*np.pi, 200)
    
    # Base ellipse
    radius = a * b / np.sqrt((b*np.cos(t))**2 + (a*np.sin(t))**2)
    
    # Add bumps
    bump_pattern = np.ones_like(t)
    for i in range(num_bumps):
        angle = i * (2*np.pi/num_bumps)
        # Smooth bump using cosine function
        bump = bumpiness * np.cos((t - angle) * num_bumps/2)**4
        bump_pattern += bump
    
    # Apply bumps
    x = center[0] + radius * np.cos(t) * bump_pattern
    y = center[1] + radius * np.sin(t) * bump_pattern
    
    # Rotate
    x_rot = center[0] + (x - center[0]) * np.cos(theta) - (y - center[1]) * np.sin(theta)
    y_rot = center[1] + (x - center[0]) * np.sin(theta) + (y - center[1]) * np.cos(theta)
    
    return x_rot, y_rot

def is_inside_shape(x, y, shape_x, shape_y):
    """Check if points (x,y) are inside a polygon"""
    polygon = np.column_stack((shape_x, shape_y))
    path = Path(polygon)
    points = np.column_stack((x.flatten(), y.flatten()))
    return path.contains_points(points).reshape(x.shape)

# Parameters you can adjust:
SPHEROID1_PARAMS = {
    'center': (-0.8, 0),      # Center position (x,y)
    'a': 0.4,              # Major axis
    'b': 0.3,              # Minor axis
    'theta': np.pi/4,      # Rotation angle
    'bumpiness': 0.15,     # Bump intensity (0 = smooth)
    'num_bumps': 12        # Number of bumps
}

SPHEROID2_PARAMS = {
    'center': (0.9, 0),      # Center position (x,y)
    'a': 0.5,              # Major axis
    'b': 0.4,              # Minor axis
    'theta': -np.pi/6,     # Rotation angle
    'bumpiness': 0.1,      # Bump intensity (0 = smooth)
    'num_bumps': 8         # Number of bumps
}

# Create two bumpy spheroids with the new parameters
x1, y1 = create_bumpy_spheroid(**SPHEROID1_PARAMS)
x2, y2 = create_bumpy_spheroid(**SPHEROID2_PARAMS)

# Create grid of points from -2 to 2
x = np.linspace(-2, 2, 50)
y = np.linspace(-2, 2, 50)
X, Y = np.meshgrid(x, y)

# Check which points are inside each shape
inside1 = is_inside_shape(X, Y, x1, y1)
inside2 = is_inside_shape(X, Y, x2, y2)

# Plotting
plt.figure(figsize=(8, 5))

# First plot the filled shapes (semi-transparent)
plt.fill(x1, y1, 'blue', alpha=0.2, label='Spheroid 1 Interior')
plt.fill(x2, y2, 'red', alpha=0.2, label='Spheroid 2 Interior')

# Then plot the grid points with different colors
plt.scatter(X[~inside1 & ~inside2], Y[~inside1 & ~inside2], color='lightgray', s=5, label='Outside')
plt.scatter(X[inside1 & ~inside2], Y[inside1 & ~inside2], color='royalblue', s=5, label='Inside Spheroid 1')
plt.scatter(X[~inside1 & inside2], Y[~inside1 & inside2], color='salmon', s=5, label='Inside Spheroid 2')
plt.scatter(X[inside1 & inside2], Y[inside1 & inside2], color='purple', s=5, label='Inside Both')

# Finally plot the shape boundaries
plt.plot(x1, y1, 'b-', linewidth=2, alpha=0.7, label='Spheroid 1 Boundary')
plt.plot(x2, y2, 'r-', linewidth=2, alpha=0.7, label='Spheroid 2 Boundary')

#plt.title('Bumpy Spheroids (-2 to 2 Range)')
#plt.xlabel('X-axis')
#plt.ylabel('Y-axis')
#plt.legend(loc='upper left')
plt.grid(True, alpha=0.3)
plt.xlim(-2, 2)
plt.ylim(-1.1, 1.1)
plt.tight_layout()
plt.savefig("figs/random_volume_est.png")
plt.show()
```

```python
def find_interior_point(boundary_x, boundary_y):
    """Find a point inside the given shape"""
    path = Path(np.column_stack((boundary_x, boundary_y)))
    # Try centroid first
    centroid = [np.mean(boundary_x), np.mean(boundary_y)]
    if path.contains_point(centroid):
        return centroid
    # If centroid fails, try random points
    for _ in range(100):
        x = np.random.uniform(min(boundary_x), max(boundary_x))
        y = np.random.uniform(min(boundary_y), max(boundary_y))
        if path.contains_point([x, y]):
            return [x, y]
    raise ValueError("Could not find interior point")

def find_boundary_intersection(start_point, direction, boundary_x, boundary_y, gap=0.05):
    """Find where a ray intersects the boundary with a small gap"""
    path = Path(np.column_stack((boundary_x, boundary_y)))
    step = 0.001  # Smaller step for more precision
    distance = 0
    max_distance = 10
    
    while distance < max_distance:
        test_point = [
            start_point[0] + distance * direction[0],
            start_point[1] + distance * direction[1]
        ]
        if not path.contains_point(test_point):
            # Return point with specified gap before boundary
            return [
                start_point[0] + (distance-gap) * direction[0],
                start_point[1] + (distance-gap) * direction[1]
            ]
        distance += step
    return [
        start_point[0] + (max_distance-gap) * direction[0],
        start_point[1] + (max_distance-gap) * direction[1]
    ]

# Find interior points
point1 = find_interior_point(x1, y1)
point2 = find_interior_point(x2, y2)

# Generate random directions
num_directions = 12
angles = np.linspace(0, 2*np.pi, num_directions, endpoint=False)
directions = np.column_stack((np.cos(angles), np.sin(angles)))

# Plotting
plt.figure(figsize=(8, 5))

# Draw shapes with semi-transparent fill
plt.fill(x1, y1, 'blue', alpha=0.2)
plt.fill(x2, y2, 'red', alpha=0.2)
plt.plot(x1, y1, 'b-', linewidth=1.5, label='Spheroid 1')
plt.plot(x2, y2, 'r-', linewidth=1.5, label='Spheroid 2')

# Plot interior points
plt.scatter(*point1, color='darkblue', s=80, zorder=10, label='Spheroid 1 Center')
plt.scatter(*point2, color='darkred', s=80, zorder=10, label='Spheroid 2 Center')

# Draw arrows for first shape (blue) with gap
for direction in directions:
    end_point = find_boundary_intersection(point1, direction, x1, y1, gap=0.03)
    plt.arrow(point1[0], point1[1], 
              end_point[0]-point1[0], end_point[1]-point1[1],
              head_width=0.03, head_length=0.05, 
              fc='blue', ec='blue', alpha=0.7, linestyle='-',
              length_includes_head=True)

# Draw arrows for second shape (red) with gap
for direction in directions:
    end_point = find_boundary_intersection(point2, direction, x2, y2, gap=0.03)
    plt.arrow(point2[0], point2[1], 
              end_point[0]-point2[0], end_point[1]-point2[1],
              head_width=0.03, head_length=0.05, 
              fc='red', ec='red', alpha=0.7, linestyle='-',
              length_includes_head=True)

#plt.title('Direction Vectors with Boundary Gap')
#plt.xlabel('X-axis')
#plt.ylabel('Y-axis')
plt.grid(True, alpha=0.2)
plt.xlim(-2, 2)
plt.ylim(-1.1, 1.1)
#plt.legend()
plt.tight_layout()
plt.savefig("figs/local_volume_est.png")
plt.show()
```

# Loss Landscape

First, the loss landscape.

```python
import torch
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from typing import List

class LossLandscape:
    def __init__(self, 
                 minima_sharp_loc=-1.0, 
                 minima_wide_loc=1.0,
                 sharp_width=0.1,
                 wide_width=0.2,
                 amplitude=1.0,
                 baseline=1.0):
        """
        Initialize a customizable loss landscape with two minima.
        """
        self.minima_sharp_loc = minima_sharp_loc
        self.minima_wide_loc = minima_wide_loc
        self.sharp_width = sharp_width
        self.wide_width = wide_width
        self.amplitude = amplitude
        self.baseline = baseline
        
    def get_minima_parameters(self):
        """Return the key parameters defining the minima"""
        return {
            'minima_sharp_loc': self.minima_sharp_loc,
            'minima_wide_loc': self.minima_wide_loc,
            'sharp_width': self.sharp_width,
            'wide_width': self.wide_width
        }
        
    def __call__(self, x):
        """Compute the loss at point(s) x"""
        gaussian_sharp = self.amplitude * torch.exp(
            -0.5 * ((x - self.minima_sharp_loc) / self.sharp_width)**2
        )
        gaussian_wide = self.amplitude * torch.exp(
            -0.5 * ((x - self.minima_wide_loc) / self.wide_width)**2
        )
        return self.baseline - gaussian_sharp - gaussian_wide
    
    def visualize(self, x_range=(-2, 2), num_points=500, savetitle=None):
        """Visualize the loss landscape"""
        x = torch.linspace(x_range[0], x_range[1], num_points)
        loss = self(x)
        
        plt.figure(figsize=(10, 5))
        plt.plot(x.numpy(), loss.numpy())
        plt.xlabel('x', fontsize = 20)
        plt.ylabel('Loss', fontsize = 20)
        plt.title('Loss Landscape', fontsize =20)
        
        plt.axvline(x=self.minima_sharp_loc, color='r', 
                   linestyle='--', alpha=0.3, label='Sharp minimum')
        plt.axvline(x=self.minima_wide_loc, color='b', 
                   linestyle='--', alpha=0.3, label='Wide minimum')
        
        plt.grid(True, alpha=0.3)
        plt.legend(fontsize = 12)
        if savetitle:
            plt.savefig(f"{savetitle}.png")
        plt.show()
    
    def visualize_guess(self, x_range=(-2, 2), num_points=500, num_scatter_points=100, savetitle=None):
        """Visualize the loss landscape with colored regions and scatter points"""
        # Create the main loss curve
        x = torch.linspace(x_range[0], x_range[1], num_points)
        loss = self(x)
        
        plt.figure(figsize=(10, 5))
        
        # Plot the loss landscape
        plt.plot(x.numpy(), loss.numpy(), 'k-', linewidth=2)
        plt.xlabel('x', fontsize=20)
        plt.ylabel('Loss', fontsize=20)
        plt.title('Loss Landscape with Minima Regions', fontsize=20)
        
        # Get current axis limits for rectangle height
        ylim = plt.ylim()
        
        # Add colored regions for minima
        plt.axvspan(self.minima_sharp_loc - self.sharp_width/2, 
                    self.minima_sharp_loc + self.sharp_width/2, 
                    color='red', alpha=0.1, label='Sharp minimum region')
        
        plt.axvspan(self.minima_wide_loc - self.wide_width/2, 
                    self.minima_wide_loc + self.wide_width/2, 
                    color='blue', alpha=0.1, label='Wide minimum region')
        
        # Add vertical lines at minima locations
        plt.axvline(x=self.minima_sharp_loc, color='r', 
                   linestyle='--', alpha=0.5)
        plt.axvline(x=self.minima_wide_loc, color='b', 
                   linestyle='--', alpha=0.5)
        
        # Generate scatter points
        scatter_x = np.linspace(x_range[0], x_range[1], num_scatter_points)
        scatter_y = np.ones(num_scatter_points)*0.5
        
        # Determine which points fall in which regions
        in_sharp = ((scatter_x >= self.minima_sharp_loc - self.sharp_width/2) & 
                    (scatter_x <= self.minima_sharp_loc + self.sharp_width/2))
        
        in_wide = ((scatter_x >= self.minima_wide_loc - self.wide_width/2) & 
                   (scatter_x <= self.minima_wide_loc + self.wide_width/2))
        
        # Plot scatter points with different colors
        plt.scatter(scatter_x[~(in_sharp | in_wide)], 
                    scatter_y[~(in_sharp | in_wide)], 
                    color='gray', alpha=0.6, label='Outside regions')
        
        plt.scatter(scatter_x[in_sharp], scatter_y[in_sharp], 
                    color='red', alpha=0.8, label='In sharp region')
        
        plt.scatter(scatter_x[in_wide], scatter_y[in_wide], 
                    color='blue', alpha=0.8, label='In wide region')
        
        plt.grid(True, alpha=0.3)
        plt.legend(fontsize=12, loc='upper right')
        
        if savetitle:
            plt.savefig(f"{savetitle}.png", bbox_inches='tight', dpi=300)
        plt.show()
        
    def random_parameter_search(self, dim: int, num_samples: int = 10000, upper = 10, low = -10):
        """
        Perform random parameter search using the class's minima parameters
        
        Args:
            dim: Dimension of parameter space
            num_samples: Number of random samples to generate
            
        Returns:
            Tuple of (sharp_count, wide_count, samples)
        """
        minima_sharp_count = 0
        minima_wide_count = 0
        samples = []
        
        for _ in range(num_samples):
            params = np.random.uniform(low=low, high=upper, size=dim)
            product = np.prod(params)
            samples.append(product)
            
            if abs(product - self.minima_sharp_loc) < self.sharp_width:
                minima_sharp_count += 1
            if abs(product - self.minima_wide_loc) < self.wide_width:
                minima_wide_count += 1
        
        return minima_sharp_count, minima_wide_count, samples
        
class NParameterModel(torch.nn.Module):
    def __init__(self, initial_values: List[float]):
        super().__init__()
        # Create N parameters from the initial values
        self.params = torch.nn.ParameterList([
            torch.nn.Parameter(torch.tensor([val], dtype=torch.float32))
            for val in initial_values
        ])
    
    def forward(self):
        # Compute product of all parameters
        x = torch.prod(torch.stack([p for p in self.params]))
        return x

    def get_parameter_values(self):
        return [p.item() for p in self.params]

# Initialize our loss landscape
minima_sharp_loc=-1.0, 
minima_wide_loc=1.0,
sharp_width=0.1,
wide_width=0.2
loss_fn = LossLandscape(minima_sharp_loc=-1.0, 
                 minima_wide_loc=1.0,
                 sharp_width=0.1,
                 wide_width=0.2)

loss_fn.visualize(savetitle="figs/LossLandscape")
```

Next, we visualize the guess and check method for estimating the volume of minima.

```python
loss_fn.visualize_guess(savetitle="figs/True_Volume1d")
```

Now, we visualize an optimization algorithm. This starts with some initial point, uses an optimization algorithm to find a minima, and then estimates the volume from within that minima. The latter bit will be a bit difficult.

```python
# Initialize the model with a starting point
initial_x = 0.0 
model = NParameterModel([initial_x])

# Create optimizer (AdamW)
optimizer = optim.AdamW(model.parameters(), lr=0.1)

# Training parameters
num_epochs = 100
x_history = []
loss_history = []

# Training loop
for epoch in range(num_epochs):
    # Zero gradients
    optimizer.zero_grad()
    x = model()
    loss = loss_fn(x)
    loss.backward()
    optimizer.step()
    
    # Store history for visualization
    x_history.append(x.item())
    loss_history.append(loss.item())
    
    # Print progress
    if (epoch + 1) % 10 == 0:
        print(f'Epoch {epoch+1}/{num_epochs}, x: {x.item():.4f}, Loss: {loss.item():.4f}')

# Plot the loss landscape and optimization path
x_vals = np.linspace(-2, 2, 500)
with torch.no_grad():
    loss_vals = loss_fn(torch.tensor(x_vals, dtype=torch.float32)).numpy()

print ("Done finding the minima!")
```

```python
params = loss_fn.get_minima_parameters()

# Plot loss landscape
plt.figure(figsize=(10, 5))
plt.plot(x_vals, loss_vals, label='Loss Landscape', linewidth=2)

# Plot optimization path
plt.scatter(x_history, loss_history, c='r', s=20, alpha=0.5, label='Optimization Path')
plt.scatter([x_history[0]], [loss_history[0]], c='g', s=100, label='Start')
plt.scatter([x_history[-1]], [loss_history[-1]], c='b', s=100, label='End')

# Plot minima boundary lines
plt.axvline(x=params['minima_sharp_loc'] - params['sharp_width']/2, 
            color='r', linestyle='--', alpha=0.5, label='Sharp Min Boundary')
plt.axvline(x=params['minima_sharp_loc'] + params['sharp_width']/2, 
            color='r', linestyle='--', alpha=0.5)

plt.axvline(x=params['minima_wide_loc'] - params['wide_width']/2, 
            color='b', linestyle='--', alpha=0.5, label='Wide Min Boundary')
plt.axvline(x=params['minima_wide_loc'] + params['wide_width']/2, 
            color='b', linestyle='--', alpha=0.5)

# Add vertical lines at minima locations
plt.axvline(x=params['minima_sharp_loc'], color='r', 
           linestyle='-', alpha=0.3, linewidth=2)
plt.axvline(x=params['minima_wide_loc'], color='b', 
           linestyle='-', alpha=0.3, linewidth=2)

# Add arrows from last point to wide minima boundaries at y=0.5
last_x = x_history[-1]
y_arrow = 0.5  # Fixed y-coordinate for arrows

# Calculate distances to each minimum's center
dist_to_sharp = abs(last_x - params['minima_sharp_loc'])
dist_to_wide = abs(last_x - params['minima_wide_loc'])

# Determine which minimum is closer
if dist_to_sharp < dist_to_wide:
    # Point to sharp minimum boundaries
    min_loc = params['minima_sharp_loc']
    min_width = params['sharp_width']
    arrow_color = 'purple'
    min_label = 'sharp'
else:
    # Point to wide minimum boundaries
    min_loc = params['minima_wide_loc']
    min_width = params['wide_width']
    arrow_color = 'purple'
    min_label = 'wide'

# Arrow to left boundary of the closest minimum
plt.annotate('', xy=(min_loc - min_width/2, y_arrow),
             xytext=(last_x, y_arrow),
             arrowprops=dict(arrowstyle='->', color=arrow_color, lw=1.5, alpha=0.9))

# Arrow to right boundary of the closest minimum
plt.annotate('', xy=(min_loc + min_width/2, y_arrow),
             xytext=(last_x, y_arrow),
             arrowprops=dict(arrowstyle='->', color=arrow_color, lw=1.5, alpha=0.9))

plt.xlabel('x', fontsize=20)
plt.ylabel('Loss', fontsize=20)
plt.title('Gradient Descent + Local Volume Estimation', fontsize=20)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.savefig("figs/LocalVolume.png", bbox_inches='tight', dpi=300)
plt.show()
```

# 2D Volume

To emphasize the role of scale invariance, we consider 2d volume estimation. The true minima for our previous problem are now given by the following plot.

```python
# Parameters for the red curve (a1, w1)
a1 = params['minima_wide_loc']
w1 = params['wide_width']/2

# Parameters for the blue curve (a2, w2)
a2 = params['minima_sharp_loc']
w2 = params['sharp_width']/2

# Plot range
x_min, x_max = -5, 5
y_min, y_max = -5, 5

# Create x values (avoid x=0)
x_pos = np.linspace(1e-10, x_max, 500)  # Positive x
x_neg = np.linspace(-x_max, -1e-10, 500)  # Negative x

# Calculate bounds for red curve (a1, w1)
red_upper_pos = (a1 + 2*w1)/x_pos
red_lower_pos = (a1 - 2*w1)/x_pos
red_upper_neg = (a1 + 2*w1)/x_neg
red_lower_neg = (a1 - 2*w1)/x_neg

# Calculate bounds for blue curve (a2, w2)
blue_upper_pos = (a2 + 2*w2)/x_pos
blue_lower_pos = (a2 - 2*w2)/x_pos
blue_upper_neg = (a2 + 2*w2)/x_neg
blue_lower_neg = (a2 - 2*w2)/x_neg

# Create the plot
plt.figure(figsize=(8, 8))

# Fill the red region (a1  2w1)
plt.fill_between(x_pos, red_lower_pos, red_upper_pos, color='red', alpha=0.3, label=f'a = {a1}  {2*w1}')
plt.fill_between(x_neg, red_lower_neg, red_upper_neg, color='red', alpha=0.3)

# Fill the blue region (a2  2w2)
plt.fill_between(x_pos, blue_lower_pos, blue_upper_pos, color='blue', alpha=0.3, label=f'a = {a2}  {2*w2}')
plt.fill_between(x_neg, blue_lower_neg, blue_upper_neg, color='blue', alpha=0.3)

# Plot the boundary lines for both regions
plt.plot(x_pos, red_upper_pos, 'r-', linewidth=1, alpha=0.5)
plt.plot(x_pos, red_lower_pos, 'r-', linewidth=1, alpha=0.5)
plt.plot(x_neg, red_upper_neg, 'r-', linewidth=1, alpha=0.5)
plt.plot(x_neg, red_lower_neg, 'r-', linewidth=1, alpha=0.5)

plt.plot(x_pos, blue_upper_pos, 'b-', linewidth=1, alpha=0.5)
plt.plot(x_pos, blue_lower_pos, 'b-', linewidth=1, alpha=0.5)
plt.plot(x_neg, blue_upper_neg, 'b-', linewidth=1, alpha=0.5)
plt.plot(x_neg, blue_lower_neg, 'b-', linewidth=1, alpha=0.5)

# Plot settings
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xlabel('x', fontsize = 30)
plt.ylabel('y', fontsize = 30)
plt.title('2D Parameter Model Minima', fontsize = 30)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend(fontsize = 20)

plt.savefig("figs/Volume2D.png")
```

```python

# Parameters for the red curve (wide minimum)
a1 = params['minima_wide_loc']
w1 = params['wide_width']/2

# Parameters for the blue curve (sharp minimum)
a2 = params['minima_sharp_loc']
w2 = params['sharp_width']/2

# Plot range (now from -2 to 2)
x_min, x_max = -2, 2
y_min, y_max = -2, 2

# Create x values (avoid x=0)
x_pos = np.linspace(1e-10, x_max, 500)  # Positive x
x_neg = np.linspace(x_min, -1e-10, 500)  # Negative x

# Calculate bounds for both regions
def calculate_bounds(x, a, w):
    """Calculate upper and lower bounds for given parameters"""
    x_inv = 1/x
    upper = (a + 2*w)*x_inv
    lower = (a - 2*w)*x_inv
    return upper, lower

red_upper_pos, red_lower_pos = calculate_bounds(x_pos, a1, w1)
red_upper_neg, red_lower_neg = calculate_bounds(x_neg, a1, w1)

blue_upper_pos, blue_lower_pos = calculate_bounds(x_pos, a2, w2)
blue_upper_neg, blue_lower_neg = calculate_bounds(x_neg, a2, w2)

# Create the plot
plt.figure(figsize=(8, 8))

# Fill the regions (now properly handling both positive and negative)
plt.fill_between(x_pos, red_lower_pos, red_upper_pos, color='red', alpha=0.3, label=f'Wide minimum (a = {a1}  {2*w1})')
plt.fill_between(x_neg, red_upper_neg, red_lower_neg, color='red', alpha=0.3)  # Note order swap for negative x

plt.fill_between(x_pos, blue_lower_pos, blue_upper_pos, color='blue', alpha=0.3, label=f'Sharp minimum (a = {a2}  {2*w2})')
plt.fill_between(x_neg, blue_upper_neg, blue_lower_neg, color='blue', alpha=0.3)  # Note order swap for negative x

# Create grid of points
grid_spacing = 0.1  # Adjust this for different point densities
x_points = np.arange(x_min, x_max + grid_spacing, grid_spacing)
y_points = np.arange(y_min, y_max + grid_spacing, grid_spacing)
xx, yy = np.meshgrid(x_points, y_points)
points = np.vstack([xx.ravel(), yy.ravel()]).T

# Classify points
colors = []
for x, y in points:
    if x == 0:
        colors.append('gray')
        continue
    
    x_inv = 1/x
    # Wide minimum (red) bounds
    red_upper = (a1 + 2*w1)*x_inv
    red_lower = (a1 - 2*w1)*x_inv
    
    # Sharp minimum (blue) bounds
    blue_upper = (a2 + 2*w2)*x_inv
    blue_lower = (a2 - 2*w2)*x_inv
    
    # Check regions (accounting for inverted bounds when x is negative)
    if x > 0:
        in_red = red_lower <= y <= red_upper
        in_blue = blue_lower <= y <= blue_upper
    else:
        in_red = red_upper <= y <= red_lower
        in_blue = blue_upper <= y <= blue_lower
    
    if in_red:
        colors.append('red')
    elif in_blue:
        colors.append('blue')
    else:
        colors.append('gray')

# Plot the grid points
plt.scatter(points[:, 0], points[:, 1], c=colors, s=20, alpha=0.7)

# Plot boundary lines
for x, upper, lower, color in [
    (x_pos, red_upper_pos, red_lower_pos, 'red'),
    (x_neg, red_upper_neg, red_lower_neg, 'red'),
    (x_pos, blue_upper_pos, blue_lower_pos, 'blue'),
    (x_neg, blue_upper_neg, blue_lower_neg, 'blue')
]:
    plt.plot(x, upper, color=color, linestyle='-', linewidth=1, alpha=0.5)
    plt.plot(x, lower, color=color, linestyle='-', linewidth=1, alpha=0.5)

# Plot settings
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xlabel('x', fontsize = 30)
plt.ylabel('y', fontsize = 30)
plt.title('Random Parameters', fontsize = 30)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend(fontsize = 20)

plt.tight_layout()
plt.savefig("figs/Volume2D_with_grid.png", dpi=300, bbox_inches='tight')
plt.show()
```
