In this notebook, we create figures that summarize the geometry of the loss landscape according to the volume hypothesis.

```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import numpy as np

def create_circle_plot(red_x, red_y, red_size, 
                      green_x, green_y, green_size, 
                      blue_x, blue_y, blue_size,
                      title="Circle Plot", figsize=(5.5, 5.5)):
    """
    Create a single square figure with circles of different colors.
    
    Parameters:
    - red_x, red_y, red_size: Lists of x positions, y positions, and sizes for red circles
    - green_x, green_y, green_size: Lists for green circles
    - blue_x, blue_y, blue_size: Lists for blue circles
    - title: Title for the plot
    - figsize: Size of the figure (default is 8x8 for a square)
    
    Returns:
    - fig, ax: Figure and axes objects
    """
    # Create figure with square aspect ratio
    fig, ax = plt.subplots(1, 1, figsize=figsize)
    
    # Plot circles
    ax.scatter(green_x, green_y, s=green_size, c='green', edgecolors='black', linewidth=1, alpha=0.7)
    ax.scatter(red_x, red_y, s=red_size, c='red', edgecolors='black', linewidth=1, alpha=0.7)
    ax.scatter(blue_x, blue_y, s=blue_size, c='blue', edgecolors='black', linewidth=1, alpha=0.7)
    
    # Set title and labels
    ax.set_title(title, fontsize=30)
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.grid(True, alpha=0.3)
    
    # Set fixed axis limits from -2 to 2
    ax.set_xlim(-1.8, 1.8)
    ax.set_ylim(-1.8, 1.8)
    
    # Set equal aspect ratio to ensure the plot is square
    ax.set_aspect('equal')
    
    # Create legend elements
    if len(blue_x) > 0:
        legend_elements = [
            Line2D([0], [0], marker='o', color='w', markerfacecolor='green', 
                   markersize=10, markeredgecolor='black', linewidth=1, label='Minima Found From Training'),
            Line2D([0], [0], marker='o', color='w', markerfacecolor='red', 
                   markersize=10, markeredgecolor='black', linewidth=1, label='Minima Not Found In Training'),
            Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', 
                   markersize=10, markeredgecolor='black', linewidth=1, label='Possible Minima From Training')
        ]
    else:
        legend_elements = [
            Line2D([0], [0], marker='o', color='w', markerfacecolor='green', 
                   markersize=10, markeredgecolor='black', linewidth=1, label='Minima Found From Training'),
            Line2D([0], [0], marker='o', color='w', markerfacecolor='red', 
                   markersize=10, markeredgecolor='black', linewidth=1, label='Minima Not Found In Training'),
        ]

    # Add legend inside the plot (upper right corner)
    ax.legend(handles=legend_elements, loc='upper right', 
              frameon=True, fancybox=True, shadow=True, fontsize=15)
    
    plt.tight_layout()
    
    return fig, ax

# Sample data for a single plot
red_x = [-1.3, 1.2, 0.6, -1.2, 1.4]
red_y = [1, 0.9, -1.4, -1.1, -0.6]
red_size = [1500, 1500, 700, 3000, 3000]

green_x = [0]
green_y = [0]
green_size = [30000]

blue_x = []#-1, -1.5, 0.8, 1.0, 0.3]
blue_y = []#-1, -0.5, -1.5, -0.2, 1.0]
blue_size = []#150, 150, 150, 150, 150]

# Create the plot
fig, ax = create_circle_plot(
    red_x, red_y, red_size,
    green_x, green_y, green_size,
    blue_x, blue_y, blue_size,
    title="Loss Landscape"
)

# Save the figure
plt.savefig("figs/strong_volume_hypothesis_loss_landscape.png", bbox_inches='tight', dpi=300, facecolor='white', edgecolor='none')
plt.show()
```
