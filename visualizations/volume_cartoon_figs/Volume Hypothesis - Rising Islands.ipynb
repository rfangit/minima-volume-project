```python
import numpy as np
import matplotlib.pyplot as plt

# --- Parameters ---
gridlines_y = [2, 5, 8]   # horizontal gridlines
x_label = "Minima Size"
y_label = "Dataset Size"

# --- Function: downward parabola ---
def downward_parabola(a=1, x_shift=0, y_shift=0, x_range=(-20, 20)):
    """
    Return (x, y) for a downward parabola y = -a*(x - x_shift)^2 + y_shift
    over the given x_range.
    """
    x = np.linspace(x_range[0], x_range[1], 800)
    y = -a * (x - x_shift)**2 + y_shift
    return x, y

# --- Plotting ---
fig, ax = plt.subplots(figsize=(8, 6))

# Gridlines
for gy in gridlines_y:
    ax.axhline(gy, color="gray", linestyle="--", linewidth=2)

# Big wide parabola (lower peak, extends across wide x)
x1, y1 = downward_parabola(a=0.2, x_shift=3, y_shift=6, x_range=(-20, 20))
ax.plot(x1, y1, label="Wide parabola (low peak)", lw = 3)

# 3 sharp parabolas (higher peaks, also wide range but appear zoomed)
x2, y2 = downward_parabola(a=20, x_shift=10, y_shift=7, x_range=(-20, 20))
ax.plot(x2, y2, label="Sharp 1", lw = 3)

x3, y3 = downward_parabola(a=80, x_shift=12, y_shift=20, x_range=(-20, 20))
ax.plot(x3, y3, label="Sharp 2", lw = 3)

x4, y4 = downward_parabola(a=18, x_shift=14, y_shift=8, x_range=(-20, 20))
ax.plot(x4, y4, label="Sharp persistent", lw = 3)

# --- Formatting ---
ax.set_xlabel(x_label, fontsize=20)
ax.set_ylabel(y_label, fontsize=20)

ax.set_xlim(-2, 16)   # zoom in to region containing all parabolas
ax.set_ylim(0, 11)    # adjust y view to show them clearly

#ax.legend()
#ax.set_title("Downward-facing Quadratics", fontsize=14)

# Remove tick labels (but keep gridlines)
ax.set_xticklabels([])
ax.set_yticklabels([])

plt.savefig("figs/rising_islands.png", dpi=300, bbox_inches="tight")
plt.show()
```
