```python
import os
import numpy as np
import matplotlib.pyplot as plt

# Ensure figs directory exists
os.makedirs("figs", exist_ok=True)

# Original data points
x_points = np.array([1, 2, 3, 4])
y_points = x_points.copy()

# Wider range to show misbehavior
x_smooth = np.linspace(0, 5, 500)

# Base straight line (good fit)
y_line = x_smooth

# Polynomial term that forces passing through the points
poly_term = (x_smooth - 1)*(x_smooth - 2)*(x_smooth - 3)*(x_smooth - 4)

# Three different bad fits
y_bad1 = x_smooth + 0.4 * poly_term
y_bad2 = x_smooth - 0.6 * poly_term
y_bad3 = x_smooth - 0.1 * poly_term  # Mild wiggle

plt.figure(figsize=(7, 5))

# Original data points
plt.scatter(x_points, y_points, color="black", zorder=5, s=60)

# Line + bad fits (now with alpha for transparency)
plt.plot(x_smooth, y_bad1, color="red", linewidth=4, linestyle="--", alpha=0.6)
plt.plot(x_smooth, y_bad2, color="red", linewidth=4, linestyle="--", alpha=0.6)
plt.plot(x_smooth, y_bad3, color="red", linewidth=4, linestyle="--", alpha=0.6)

plt.plot(x_smooth, y_line, color="limegreen", linewidth=5)

plt.xlim(0, 5)
plt.ylim(0, 5)
plt.xlabel("X", fontsize=20)

# Rotate Y label horizontally for readability
plt.ylabel("Y", fontsize=20, rotation=0, labelpad=20)

# Disable grid and ticks
plt.grid(False)
plt.xticks([])
plt.yticks([])

# Save figure
save_path = "figs/simple_model.png"
plt.savefig(save_path, dpi=200, bbox_inches = "tight")

# Display the figure
plt.show()

print(f"Figure saved to {save_path}")
```
